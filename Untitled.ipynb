```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
t = np.linspace(0,1,100)
std = 1 + 0.9*np.sin(2*np.pi*t)
locs=np.zeros(len(t)) + 0.3
x = np.zeros(len(t))
for i in range(len(t)):
    x[i] = np.random.normal(loc=locs[i],scale=std[i])
plt.plot(t,x)
```

```python
def fZeroTest(x,std):
    logPs = []
    for i in range(1000):
        logP = 0
        for j in range(len(x)):
            s = np.random.normal(loc=0,scale=std[j])
            logP += -s**2/(2*std[j]**2) - np.log(std[j])
        logPs.append(logP)
    logPx = 0
    for j in range(len(x)):
        logPx += -x[j]**2/(2*std[j]**2) - np.log(std[j])
    plt.hist(logPs)
    print(logPx)
```

```python
fZeroTest(x,std)
```
