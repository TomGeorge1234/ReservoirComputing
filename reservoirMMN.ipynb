```python
inputParams = defaultInputParams.copy()
inputParams['syllables'] = alphabet[:3]
inputParams['chunkList'] = [['a']]
inputParams['gapRange'] = [0,0]
inputs_A = getInputs(inputParams,totalTime=50)
inputParams['chunkList'] = [['a']*10]
inputParams['gapRange'] = [2,2]
inputParams['interChunkSyllables'] = ['b','c',' ']
inputsTest_A = getInputs(inputParams,totalTime=50)
plotInputs(inputs_A,title='A')
plotInputs(inputsTest_A,title='A (test)')

hyperparams_A = defaultHyperparams.copy()
hyperparams_A['Nin'] = inputs_A['data'].shape[1]
hyperparams_A['Nres_out'] = 100
hyperparams_A['Nres_in'] = 100
```

```python
reservoirPair_A = ReservoirPair(hyperparams_A,inputs_A)
reservoirPair_A.storeInputs(inputsTest_A,name='test')
reservoirPair_A.trainPair(saveTrain=True,returnItems=['z'])
reservoirPair_A.testPair(testName='posttraining',testData='test',Ntest=5,returnItems=['z'])

plotTest(reservoirPair_A,testName='train',testData='train',tstart=0,tend=50,smoothedMean=True)
plotInputs(reservoirPair_A.inputDict['test'])
plotTest(reservoirPair_A,testName='posttraining',testData='test')
```
